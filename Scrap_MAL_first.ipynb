```python

import requests
from bs4 import BeautifulSoup

import csv
import lxml

def ScanPageForAnime(starting_index):
    # base page showing 50 anime listings:
    result = requests.get("https://myanimelist.net/topanime.php?limit={}".format(starting_index))

    # page HTML content:
    src = result.content

    # parse src using BeautifulSoup:
    soup = BeautifulSoup(src, 'lxml')

    # Now that the page source has been processed via Beautifulsoup
    # we can access specific information directly from it. For instance,
    # say we want to see a list of all of the links on the page:
    anime_block_list = soup.find_all("td",attrs={"class","title al va-t word-break"})

    anime_list = []
    for anime_block in anime_block_list:
        anime_link = anime_block.find("a").attrs["href"]
        anime_id, anime_name = anime_link.replace("https://myanimelist.net/anime/", "").split("/")
        anime_list.append((anime_id, anime_name))
        
    return anime_list

anime_list = []
for i in range(0, 200):
    ani = ScanPageForAnime(i*50)
    anime_list += ani
    
print(anime_list)

"""data = ScanPageForAnime(0)
with open('ur file.csv','w') as out:
    csv_out=csv.writer(out)
    csv_out.writerow(['name','num'])
    for row in data:
        csv_out.writerow(row)
        """ 
```
